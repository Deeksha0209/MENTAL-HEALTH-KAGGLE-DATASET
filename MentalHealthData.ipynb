```python
#Creating a bar plot for cateforical values using color map and functions!
color_map = {
    'Yes': 'green',
    'No': 'red',
    'Maybe': 'pink',
    'Not sure': 'pink'
}
```

```python
def bar_plot(variable):
    var = df[variable]
    Count = var.value_counts()

    # Map colors based on category name
    bar_colors = [color_map.get(cat, 'gray') for cat in Count.index]

    # Plot
    plt.figure(figsize=(9, 3))
    plt.bar(Count.index, Count, color=bar_colors)
    plt.xticks(rotation=0)
    plt.ylabel("Count")
    plt.title(variable)
    plt.show()

    print(f"{variable}: \n{Count}")
```

```python
category = ["FamilyHistory", "Treatment", "MentalHealthHistory", "MentalHealthInterview",
            "CareOptions"]

for c in category:
    bar_plot(c)
```

```python
import plotly.express as px

# Filter dataset
filtered_df = df[
    (df['Gender'].str.strip().str.lower() == 'male') &
    (df['SelfEmployed'].astype(str).str.strip().str.lower() == 'yes')
]

cols = ['DaysIndoors', 'MoodSwings', 'SocialWeakness', 'HabitsChange']
filtered_df = filtered_df[cols].dropna()

# Create treemap
fig = px.treemap(
    filtered_df,
    path=cols,      # hierarchy of categories
    color='MoodSwings',  # optional: color by one column
    color_discrete_map={'Yes':'green', 'No':'blue', 'Maybe':'yellow'},  # define colors
    title='Self-Employed Males: Mosaic Interactive Plot' ,
    hover_data=cols  # show all categories on hover
)

# Customize hover template for clarity
fig.update_traces(
    hovertemplate='<br>'.join([f'{col}: %{{customdata[{i}]}}' for i, col in enumerate(cols)]) +
                  '<br>Count: %{value}<extra></extra>'
)

fig.show()
```

```python
countries = ['usa', 'canada', 'uk']  # list of countries (lowercase)
country_df = df[df['Country'].str.strip().str.lower().isin(countries)]


cols = ['Country', 'CareOptions']
bar_df = country_df[cols].dropna()

# Simple bar chart
plt.figure(figsize=(12, 6))
sns.countplot(data=bar_df, hue='Country', x='CareOptions', palette='magma', orient='h')
plt.title('Country vs Care Options')
plt.ylabel('Count')
plt.legend(title='Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cols = ['Gender', 'MentalHealthHistory', 'Treatment', 'CareOptions']
heatmap_df = df[cols]

# Create a grouped count. Groups your data by the combination of the four categorical columns. size() Counts how many rows are in each group &
#reset_index() Converts the resulting Series into a DataFrame and names the count column "count"
agg_df = heatmap_df.groupby(['Gender', 'MentalHealthHistory', 'Treatment', 'CareOptions']).size().reset_index(name='count')
#Next, we loop over unique values for treatment and care options.
for treatment in agg_df['Treatment'].unique():
    for care in agg_df['CareOptions'].unique():
#Nested loop filters data for every combination of treatment and care. eg. T=Y, C=Y or T=Y, C=N or T=N, C=N and T=N, C=Y.
        subset = agg_df[(agg_df['Treatment'] == treatment) & (agg_df['CareOptions'] == care)]
        pivot_df = subset.pivot(index='Gender', columns='MentalHealthHistory', values='count').fillna(0)
#Here, rows will include Male, Female and columns = Mental Health History count
        plt.figure(figsize=(6, 4))
        sns.heatmap(pivot_df, annot=True, fmt="g", cmap="flare")
        plt.title(f'Heatmap: Treatment={treatment}, CareOptions={care}')
        plt.ylabel('Gender')
        plt.xlabel('MentalHealthHistory')
        plt.show()
```
